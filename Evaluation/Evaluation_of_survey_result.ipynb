#SURVEY 1

```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

labels1 = ['1', '2', '3', '4', '5']
sizes1 = [2.7, 5.4, 8.1, 56.8, 27]
explode1 = (0, 0, 0, 0.1, 0)

labels2 = ['1', '2', '3', '4', '5']
sizes2 = [5.4, 2.7, 16.2, 40.5, 35.1]
explode2 = (0, 0, 0, 0.1, 0)

labels3 = ['1', '2', '3', '4', '5']
sizes3 = [2.7, 8.1, 16.2, 43.2, 29.7]
explode3 = (0, 0, 0, 0.1, 0)

labels4 = ['1', '2', '3', '4', '5']
sizes4 = [21.6, 18.9, 21.6, 24.3, 13.5]
explode4 = (0, 0, 0, 0.1, 0)

labels5 = ['1', '2', '3', '4', '5']   
sizes5 = [10.8, 29.7, 21.6, 24.3, 13.5]
explode5 = (0, 0.1, 0, 0, 0)

sns.set(style="darkgrid")
fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(ncols=5, figsize=(15, 4))

ax1.pie(sizes1, explode=explode1, labels=labels1, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
ax1.title.set_text('Black Hair')

ax2.pie(sizes2, explode=explode2, labels=labels2, autopct='%1.1f%%', shadow=True, startangle=90)
ax2.axis('equal')
ax2.title.set_text('Blonde Hair')

ax3.pie(sizes3, explode=explode3, labels=labels3, autopct='%1.1f%%', shadow=True, startangle=90)
ax3.axis('equal')  
ax3.title.set_text('Brown Hair')

ax4.pie(sizes4, explode=explode4, labels=labels4, autopct='%1.1f%%', shadow=True, startangle=90)
ax4.axis('equal')
ax4.title.set_text('Gender')

ax5.pie(sizes5, explode=explode5, labels=labels5, autopct='%1.1f%%', shadow=True, startangle=90)
ax5.axis('equal')  
ax5.title.set_text('Age')

plt.tight_layout()
plt.show()
```

```
y_pos = ('Blonde Hair', 'Gender', 'Age', 'Brown Hair', 'Black Hair')
det_acc = [37.8, 24.3, 21.6, 18.9, 16.2]

sns.set(style="darkgrid")
plt.figure(figsize=(7.5,7.5))
plt.bar(y_pos, det_acc, align='center', alpha=1, color='black')
plt.ylim(0,40)
plt.ylabel('Percentages')
plt.title('% of respondents accurately determining transformed image', size=13, y=1.025)

plt.show()
```
